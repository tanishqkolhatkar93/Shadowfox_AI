# 7. Create a LLM Wrapper

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
from langchain import HuggingFacePipeline
import torch
```

### HuggingFace Notebook Login (OPEN SOURCE)

```python
from huggingface_hub import notebook_login
```

```python
# Copy and paste your hhuggingface hub token and login
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True,)
```

```python
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
```

```python
pipe = transformers.pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0.01})
```

```python
llm.predict("Please provide a concise summary of the Book, Game of Thrones")
```

# 8. Initialize the RetrievalQA Chain

```python
from langchain.chains import RetrievalQA
```

### Base retriever

```python
 # create the chain to answer questions
qa_chain = RetrievalQA.from_chain_type(llm=llm,
                                       chain_type="stuff",
                                       retriever=docsearch.as_retriever(search_kwargs={'k': 2}),
                                       return_source_documents=False)
```

```python
user_input = "What is RAG"
```

```python
print(qa_chain.run(user_input))
```
